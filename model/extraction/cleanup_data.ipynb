### Initial data cleanup

```python
import pandas as pd

RAW_CSV_FILEPATH = '../data/raw_scraped_car_listings.csv'
CLEANED_CSV_FILEPATH = '../data/cleaned_car_listings.csv'
```

```python
df = pd.read_csv(RAW_CSV_FILEPATH)
df
```

```python
df['url'].nunique()
```

The **url** column was used to prevent scraping the same listing page twice, it's not a relevant feature, so we'll discard it.

```python
# drop the url column
df = df.drop(columns=['url'])
df
```

```python
df.info()
```

In the data collection process, the listings that did not contain values for the selected features were ignored, so we don't have any missing values in the dataset.

```python
# display missing values count
df.isnull().sum()
```

#### Data cleanup

For the **categorical columns** cleanup the strings (remove trailing whitespaces):

```python
categorical_columns = [
    "make",
    "model",
    "body_type",
    "color",
    "fuel_type",
    "gearbox",
    "transmission",
    "pollution_standard"
]
```

```python
def cleanup_string(string):
    # remove trailing whitespaces
    string = ' '.join(string.split())

    return string

for col in categorical_columns:
    df[col] = df[col].apply(cleanup_string)

df
```

For all numerical columns (year, door_count, nr_seats, engine_capacity, engine_power, price), remove whitespace, other string formats (km, cm3, CP) and convert to float type.

For `mileage`, convert from string format "194 382 km" to numerical value, remove "km".

For `engine_capacity`, convert from string format "1 995 cm3" to numerical value, remove "cm3".

For `engine_power`, convert from string format "190 CP" to numerical value, remove "CP".

For `price`, convert commas ',' to decimal points '.'.

```python
# check that all entries have a value in the mileage column that matches the pattern "number km"
matches = df['mileage'].str.match(r'^\d+(\s?\d+)*\s+km$')
print(f'[{matches.sum()}]', "All values from the mileage column respect the format <number km>"if matches.sum() == len(df) else "Unexpected format in mileage column")

# check that all entries have a value in the engine_capacity column that matches the pattern "number cm3"
matches = df['engine_capacity'].str.match(r'^\d+(\s?\d+)*\s+cm3$')
print(f'[{matches.sum()}]', "All values from the engine_capacity column respect the format <number cm3>"if matches.sum() == len(df) else "Unexpected format in engine_capacity column")

# check that all entries have a value in the engine_power column that matches the pattern "number CP"
matches = df['engine_power'].str.match(r'^\d+(\s?\d+)*\s+CP$')
print(f'[{matches.sum()}]', "All values from the engine_power column respect the format <number CP>"if matches.sum() == len(df) else "Unexpected format in engine_power column")
```

```python
df['mileage'] = df['mileage'].str.replace('km', '')
df['engine_capacity'] = df['engine_capacity'].str.replace('cm3', '')
df['engine_power'] = df['engine_power'].str.replace('CP', '')
df['price'] = df['price'].str.replace(',','.')
df
```

For all numerical columns, remove whitespace and convert to float type.

```python
numerical_columns = [
    "year",
    "mileage",
    "door_count",
    "nr_seats",
    "engine_capacity",
    "engine_power",
    "price"
]

for col in numerical_columns:
    if df[col].dtype == 'object':
        df[col] = df[col].str.replace(' ', '')
    df[col] = df[col].astype(float)

df
```
